```python
import numpy as np
import matplotlib.pyplot as plt

#define the analog signal...
t = np.arange(0,11,1)
x = (0.85)**t
plt.figure(figsize=(10,8))
plt.subplot(2,2,1)
plt.plot(t,x,label='x(t)=0.85^t')
plt.title('Analog signal')
plt.xlabel('t')
plt.ylabel('Amplitude')
plt.legend(loc = 'upper right')


#sampling and plotting of sampled signal 
plt.subplot(2,2,2)
plt.title('sampling')
plt.plot(t,x,label = 'x(t)= 0.85^t')

tt = np.linspace(0,10,50)
xx = (0.85)**tt
plt.stem (tt,xx,label = 'x(n) = 0.85^n')
plt.xlabel('n')
plt.ylabel('amplitude')
plt.legend()

#quantization 
plt.subplot(2,2,3)
plt.plot(t,x)
plt.stem(tt,xx)
plt.title('quantizaion')
plt.xlabel('n')
plt.ylabel('range of quantizer')

#ploting horizontal line accross the axes of the plot
plt.axhline(y=0.1,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.2,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.3,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.4,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.5,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.6,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.7,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.8,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.9,xmin=0,xmax = 10,color='r')
plt.axhline(y=1.0,xmin=0,xmax = 10,color='r')

#quantize signal
plt.subplot(2,2,4)
plt.title('quantize signal')
xq = np.around(xx,1) #rounding each element to 1 decemial places
                     #we have taken rounding one decimal places causes we divide the y-axis into 10 part
                     #if we take 2 decimal placese then y-axis divided = (10^2) = 100
                     #if we take 1 decimal places then y-axis divided  = (10^1) = 10
plt.stem(tt,xq)
plt.xlabel('n')
plt.ylabel('range of quantizer')

#ploting horizontal line across the Axes of the plot
plt.axhline(y=0.1,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.2,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.3,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.4,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.5,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.6,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.7,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.8,xmin=0,xmax = 10,color='r')
plt.axhline(y=0.9,xmin=0,xmax = 10,color='r')
plt.axhline(y=1.0,xmin=0,xmax = 10,color='r')
plt.tight_layout()
plt.show()
print(tt)
print(xx)
print(xq)
```
