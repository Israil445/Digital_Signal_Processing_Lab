```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
#analog signal
f1 = 1000
f2 = 2000
ph2 = (3*np.pi)/4
sampling_rate  = 8000
time = np.linspace(0,0.001,sampling_rate)
signal = np.sin(2*np.pi*f1*time) + 0.5*np.sin((2*np.pi*f2*time)+ph2)

plt.plot(time,signal,label = 'analog signal')
plt.xlabel('time')
plt.ylabel('amplitude')
plt.title('analog signal')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#sampled signal
x = []
ts = 1/sampling_rate

for i in range(0,8):
    sg = np.sin(2*np.pi*f1*i*ts) + 0.5*np.sin(2*np.pi*f2*i*ts + ph2)
    sg = float("{:.3f}".format(sg))
    x.append(sg)

index = np.arange(8)
plt.stem(index,x)
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('sampled signal')
print(x)
    
```

```python
#calculate real and imaginary part of X(m)
#X(m) = X_real + X_imag

X_real = []
X_img  = []

for m in range(0,8):
    a = 0
    b = 0
    for n in range(0,8):
        a+=(x[n]*(np.cos((2*np.pi*n*m)/8)))
        a = float("{:.3f}".format(a))
        b-=(x[n]*(np.sin((2*np.pi*n*m)/8)))
        b = float("{:.3f}".format(b))

    X_real.append(a)
    X_img.append(b)

print(X_real)
print(X_img)
```

```python
#ploting the real part 
plt.stem(np.arange(0,8),X_real)
plt.xlabel('m')
plt.ylabel('amplitude')
plt.title('real part of X(m)')
```

```python
#ploting the imaginary part 
plt.stem(np.arange(0,8),X_img)
plt.xlabel('m')
plt.ylabel('amplitude')
plt.title('imaginary part of X(m)')
```

```python
#ploting the magnitude sepcturm
Xm = []
for i in range(0,8):
    a = np.sqrt(X_real[i]**2 + X_img[i]**2)
    Xm.append(a)

index = np.arange(0,8)
plt.stem(index,Xm)
plt.xlabel('m')
plt.ylabel('magnitude')
plt.title('magnitude of X(m)')
```

```python
#ploting the power spectrum
Xm = []
for i in range(0,8):
    a = (X_real[i]**2 + X_img[i]**2)
    Xm.append(a)

index = np.arange(0,8)
plt.stem(index,Xm)
plt.xlabel('m')
plt.ylabel('magnitude')
plt.title('power spectrum of X(m)')
```

```python
#phase specturm
phase = []
for i in range(0,8):
    angle = math.atan2(X_img[i],X_real[i])
    angle = math.degrees(angle)
    if angle == 180:
        angle = 0

    phase.append(angle)

plt.stem(np.arange(0,8),phase)
plt.xlabel('m')
plt.ylabel('amplitude in degrees')
plt.title('phase spectrum')
print(phase)
```

```python
#inverse DFT 

x_real = []
x_img = []

for n in range(0,8):
    a = 0
    b = 0
    for m in range(0,8):
        a += (X_real[m]* np.cos(2 * np.pi * n * m / 8))-(X_img[m]* np.sin(2 * np.pi * n * m / 8))
        a = float("{:.3f}".format(a))
        b += (X_real[m]* np.sin(2 * np.pi * n * m / 8))+(X_img[m]*np.cos(2 * np.pi * n * m / 8))
        b = float("{:.3f}".format(b))
    a=a/8
    b=b/8
    x_real.append(a)
    x_img.append(b)
print(x_real)
print(x_img)
print(x)
plt.stem(np.arange(8),x_real)
plt.show()
```
