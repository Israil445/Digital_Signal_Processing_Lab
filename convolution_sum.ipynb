```python
import numpy as np
import matplotlib.pyplot as plt
#calculating convolution sum
def convolution_sum(x,h):
    y = {key: 0  for key in range(-100,101)}
    for n in range(-100,101):
        for k in range(-100,101):
            if -100<=n-k<=100:
                y[n]+=(x[k]*h[n-k])
    return y;
    
#starting here
#set x and h default value 0 with length 100
x = {key: 0 for key in range(-100,101)}#dictionary in python
h = {key: 0 for key in range(-100,101)}#dictionary in python

#define x and h 
x[0]=3
x[1]=2
x[2]=1
x[3]=2

h[-1]=1
h[0]=2
h[1]=1
h[2]=2

y = convolution_sum(x,h)
#method-1 for extracting data values from the y array
x_low = 0
h_low = -1
n1 = 4 #length of the x(n)
n2 = 4 #length of the h(n)

st_indx = x_low + h_low
length =  n1 + n2 - 1
result = []
for i in range(st_indx,st_indx+length):
    result.append(y[i])
print(result)

#method-2 for extracting data values from the y array
#ploting the convolution sum
convolution_result = []
index = []

for i in range(-100,101):
    if y[i]!=0:
        convolution_result.append(y[i])
        index.append(i)

print("index:",index)
print("convolution sum",convolution_result)
plt.stem(index, convolution_result)
plt.xlabel('index')
plt.ylabel('y(n)')
plt.title('convolution sum')
plt.grid(True)
plt.show()
        
```
