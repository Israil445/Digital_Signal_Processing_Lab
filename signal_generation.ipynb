```python
import numpy as np
import matplotlib.pyplot as plt

#sin and cosine  wave signal generation
t = np.arange(0,np.pi,0.001)
f = 1
y = np.sin(2*np.pi*f*t)
y1 = np.cos(2*np.pi*f*t)
print(len(t))
plt.plot(t,y,label='sin wave')
plt.plot(t,y1,label='cos wave')
plt.ylabel('Amplitude')
plt.xlabel('t')
plt.title('sinusodial signal')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#generating some elementary signals...

#unit sample sequence or unit impulse
x = np.zeros(20)
n = np.arange(-10,10)
#set the value of the first element to 1
x[10] = 1;
plt.subplot(2,2,1)
plt.stem(n,x,label='unit sample sequence')
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('unit sample signal')
plt.grid(True)
plt.legend(loc = 'upper right' )
# plt.show()

#unit step signal...
x = np.ones(10)
n = np.arange(0,10)
plt.subplot(2,2,2)
plt.stem(n, x,label = 'unit step signal')
plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Unit Step Sequence')
plt.grid(True)
plt.legend(loc = 'upper right')
# plt.show()

#unit ramp signal 
x = 30
t = np.arange(0,x + 1)
print(n)
print(x)
#ploting unit ramp signal
plt.subplot(2,2,3)
plt.stem(t,t)
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.title('Ramp Sequence')
plt.grid(True)

#exponential signal
n = np.arange(0,21)
x2 = 0.8 ** n
#ploting the exponential signal
plt.subplot(2,2,4)
plt.stem(n, x2,label = 'exponential signal')

plt.xlabel('n')
plt.ylabel('x(n)')
plt.title('Exponential Sequence')
plt.grid(True)
plt.tight_layout()
plt.show()
```
