```python
import numpy as np
import matplotlib.pyplot as plt
import math

t = np.linspace(0,512,512)
x1 = 10*np.sin(2*np.pi*40*t)
x2 = 20*np.sin(2*np.pi*80*t)
x3 = 40*np.sin(2*np.pi*160*t)

x = x1+x2+x3
plt.plot(t,x)
plt.xlabel('t')
plt.ylabel('amplitude')
plt.title('composite signal')
plt.grid(True)
plt.show()
print(len(x))
```

```python
fs = 1/512
ts = np.arange(0,0.5,fs)
print(len(ts))

x1s = 10*np.sin(2*np.pi*(40)*ts)
x2s = 20*np.sin(2*np.pi*(80)*ts)
x3s = 40*np.sin(2*np.pi*(160)*ts)

xs = x1s + x2s + x3s
plt.plot(ts,xs)
#plt.stem(ts,xs)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.title('sampled signal')
plt.grid(True)
plt.show()
```

```python
#down sampling or under sampling
fs = 1/150
ts = np.arange(0,0.5,fs)

x1s = 10*np.sin(2*np.pi*(40)*ts)
x2s = 20*np.sin(2*np.pi*(80)*ts)
x3s = 40*np.sin(2*np.pi*(160)*ts)

xs = x1s + x2s + x3s
plt.plot(ts,xs)
#plt.stem(ts,xs)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.title('down sampling')
plt.grid(True)
plt.show()
```

```python
#up sampling or over sampling
fs = 1/(512*4)
ts = np.arange(0,0.5,fs)

x1s = 10*np.sin(2*np.pi*(40)*ts)
x2s = 20*np.sin(2*np.pi*(80)*ts)
x3s = 40*np.sin(2*np.pi*(160)*ts)

xs = x1s + x2s + x3s
plt.plot(ts,xs)
#plt.stem(ts,xs)
plt.xlabel('n')
plt.ylabel('amplitude')
plt.title('up sampling')
plt.grid(True)
plt.show()
```
